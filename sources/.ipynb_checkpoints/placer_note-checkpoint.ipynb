```python
from placer import *
```

```python
im = get_images('../example_test')
```

```python
for image in im:
    img, box, rect = find_poly(image)
    x,y,w,h = box
    cv2.rectangle(img,(x,y),(x+w,y+h),(0,255,0),30)
    plt.imshow(img)
    plt.show()
    plt.imshow(rect, cmap='gray')
    plt.show()
    
    wf = without_figures(image, box)
    plt.imshow(without_figures(image, box), cmap='gray')
    plt.show()
    masks = get_masks(wf)
    
    for msk in masks:
        plt.imshow(msk)
        plt.show()
```
